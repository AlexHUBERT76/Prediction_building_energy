# Predict the energy consumption of buildings (3)

## Here we implement our model to predict the emision of future buildings

We will compare different algorythm : dummy for benchmarking, linear, lasso, ridge, random forest and XGBoost. We will rank them using R2 coeficient. We will then compare our best model but using the energy star score features and conclude on it's usefullness.

First we import our clean data :

```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.dummy import DummyRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, median_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
import optuna
import xgboost as xgb



pd.set_option("display.max_rows" , 500)
pd.set_option("display.max_columns" , 500)
rawdf = pd.read_csv(r'E:\DL\2016_Building_Energy_Benchmarking_cleaned.csv', encoding = 'utf8',sep=',',low_memory=False) #load csv file
rawdf = rawdf.set_index(keys="OSEBuildingID")
print("Dataset has ",rawdf.shape[0]," individuals and ",rawdf.shape[1]," variables.")

df = rawdf.drop(columns=["ENERGYSTARScore"])

#Dataframe for results
resultdf = pd.DataFrame(columns=['Model', 'R2', 'RMSE', 'MAE', 'Train_time', 'Test_time'])

splitrandomstate = 1 #random state for value splitting

```

Splitting train and test values :

```python
#splitting train and test values
train, test = train_test_split(df, test_size=0.2,random_state = splitrandomstate)

y_train = train["TotalGHGEmissions"].values
y_test = test["TotalGHGEmissions"].values

train = train.drop(columns = ["SiteEnergyUseWN(kBtu)","TotalGHGEmissions"])
test = test.drop(columns = ["SiteEnergyUseWN(kBtu)","TotalGHGEmissions"])

X_train = train.values
X_test = test.values
```

Implementing dummy model and linear model :

```python
starttime = time.time()
#add random state here
lm = LinearRegression().fit(X_train, y_train)
fittime1 = time.time()-starttime
            
starttime = time.time()
lm_dummy_mean = DummyRegressor(strategy = 'mean').fit(X_train, y_train)  
fittime2 = time.time()-starttime
            
starttime = time.time()
lm_dummy_median = DummyRegressor(strategy = 'median').fit(X_train, y_train)
fittime3 = time.time()-starttime

 
starttime = time.time()            
y_predict = lm.predict(X_test)
testtime1 = (time.time()-starttime)
            
starttime = time.time()
y_predict_dummy_mean = lm_dummy_mean.predict(X_test)
testtime2 = (time.time()-starttime)
            
starttime = time.time()
y_predict_dummy_median = lm_dummy_median.predict(X_test)
testtime3 = (time.time()-starttime)

r2 = r2_score(y_test, y_predict_dummy_mean)
rmse = np.sqrt(mean_squared_error(y_test, y_predict_dummy_mean))
mae = median_absolute_error(y_test, y_predict_dummy_mean)

new_row = pd.Series({'Model':DummyRegressor(strategy = 'mean'), 'R2':r2, 'RMSE':rmse, 'MAE':mae,'Train_time': fittime1, 'Test_time': testtime1})
resultdf = pd.concat([resultdf, new_row.to_frame().T], ignore_index=True)

r2 = r2_score(y_test, y_predict_dummy_median)
rmse = np.sqrt(mean_squared_error(y_test, y_predict_dummy_median))
mae = median_absolute_error(y_test, y_predict_dummy_median)

new_row = pd.Series({'Model':DummyRegressor(strategy = 'median'), 'R2':r2, 'RMSE':rmse, 'MAE':mae,'Train_time': fittime2, 'Test_time': testtime2})
resultdf = pd.concat([resultdf, new_row.to_frame().T], ignore_index=True)


r2 = r2_score(y_test, y_predict)
rmse = np.sqrt(mean_squared_error(y_test, y_predict))
mae = median_absolute_error(y_test, y_predict)

new_row = pd.Series({'Model':LinearRegression(), 'R2':r2, 'RMSE':rmse, 'MAE':mae,'Train_time': fittime3, 'Test_time': testtime3})
resultdf = pd.concat([resultdf, new_row.to_frame().T], ignore_index=True)
```

```python
#plotting accuracy

plt.scatter(y_test, y_predict, c='r')
plt.xlabel('True values')
plt.ylabel('Predicted values')

xpoints = plt.xlim()
ypoints = plt.ylim()
plt.plot(xpoints, ypoints, 'k--', alpha=.8, scalex=False, scaley=False)

plt.grid()



plt.title('Model accuracy', fontsize=15)
plt.show()
```

Implementing ridge model and tunning the hyperparameter with a cross validated grid search :

```python


param_number = 1000
alphas = np.linspace(0.1,5.,num = param_number)
parameters = {'alpha': alphas}

# define the model/ estimator
model = Ridge()

# define the grid search
Ridge_reg= GridSearchCV(model, parameters, scoring="r2",cv=5)

starttime = time.time()

#fit the grid search
Ridge_reg.fit(X_train,y_train)

fittime = (time.time()-starttime)

# best model
best_model = Ridge_reg.best_estimator_
best_model.fit(X_train,y_train)

starttime = time.time()
y_predict = best_model.predict(X_test)
testtime = (time.time()-starttime)

r2 = r2_score(y_test, y_predict)
rmse = np.sqrt(mean_squared_error(y_test, y_predict))
mae = median_absolute_error(y_test, y_predict)

new_row = pd.Series({'Model':Ridge_reg.best_estimator_, 'R2':r2, 'RMSE':rmse, 'MAE':mae,'Train_time': fittime, 'Test_time': testtime})
resultdf = pd.concat([resultdf, new_row.to_frame().T], ignore_index=True)
resultdf
```

Implementing lasso model and tunning the hyperparameter with a cross validated grid search :

```python


param_number = 1000
alphas = np.linspace(0.00001,0.01,num = param_number)
parameters = {'alpha': alphas}

# define the model/ estimator
model = Lasso()

# define the grid search
Lasso_reg= GridSearchCV(model, parameters, scoring="r2",cv=5)

starttime = time.time()

#fit the grid search
Lasso_reg.fit(X_train,y_train)

fittime = (time.time()-starttime)

# best model
best_model = Lasso_reg.best_estimator_
best_model.fit(X_train,y_train)

starttime = time.time()
y_predict = best_model.predict(X_test)
testtime = (time.time()-starttime)

r2 = r2_score(y_test, y_predict)
rmse = np.sqrt(mean_squared_error(y_test, y_predict))
mae = median_absolute_error(y_test, y_predict)

new_row = pd.Series({'Model':best_model, 'R2':r2, 'RMSE':rmse, 'MAE':mae,'Train_time': fittime, 'Test_time': testtime})
resultdf = pd.concat([resultdf, new_row.to_frame().T], ignore_index=True)
resultdf
```

Implementing random forest model and tunning the hyperparameters with a cross validated grid search :

```python
# Create the parameter grid
param_grid = {
    'bootstrap': [True],
    'max_depth': [80,90,100],
    'max_features': [9,18,19,20],
    'min_samples_leaf': [1],
    'min_samples_split': [4,5,6],
    'n_estimators': [100,550,600,650]
}
# Create a based model
rf = RandomForestRegressor()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 5, n_jobs = -1, verbose = 2, scoring="r2")

starttime = time.time()

#fit the grid search
grid_search.fit(X_train,y_train)

fittime = (time.time()-starttime)

# best model
best_model = grid_search.best_estimator_
best_model.fit(X_train,y_train)

starttime = time.time()
y_predict = best_model.predict(X_test)
testtime = (time.time()-starttime)

r2 = r2_score(y_test, y_predict)
rmse = np.sqrt(mean_squared_error(y_test, y_predict))
mae = median_absolute_error(y_test, y_predict)

new_row = pd.Series({'Model':"RandomForest", 'R2':r2, 'RMSE':rmse, 'MAE':mae,'Train_time': fittime, 'Test_time': testtime})
resultdf = pd.concat([resultdf, new_row.to_frame().T], ignore_index=True)
```

```python
dtrain = xgb.DMatrix(data = X_train, label = y_train)
dtest = xgb.DMatrix(data = X_test, label = y_test)

def objective(n_trials):
        
    params = {
        "learning_rate": n_trials.suggest_float("learning_rate", 0, 1),
        "min_split_loss": n_trials.suggest_float("min_split_loss", 0, 10),
        "reg_lambda": n_trials.suggest_float("reg_lambda", 0,1),
        "reg_alpha": n_trials.suggest_float("reg_alpha", 0,1),
        "max_depth": n_trials.suggest_int("max_depth", 2, 15),
        "colsample_bytree": n_trials.suggest_float("colsample_bytree", 0,1),
    }


    # Use cross-validation to calculate mean r2 score
    cv_results = xgb.cv(
        params, dtrain, num_boost_round=100, nfold=3, stratified=False, metrics="rmse",
        as_pandas=True, seed=123, early_stopping_rounds=50
    )
    
    r2 = -cv_results["test-rmse-mean"].iloc[-1]  # Negate rmse to maximize r2 (R^2 = 1 - (RMSE^2 / Var(y))where Var(y) is the variance of the true target values.)

    return r2
```

```python
#plotting accuracy

plt.scatter(y_test, y_predict, c='r')
plt.xlabel('True values')
plt.ylabel('Predicted values')

xpoints = plt.xlim()
ypoints = plt.ylim()
plt.plot(xpoints, ypoints, 'k--', alpha=.8, scalex=False, scaley=False)

plt.grid()



plt.title('Model accuracy', fontsize=15)
plt.show()
```

Implementing XGBoost model and tunning the hyperparameters with optuna package:

```python
# make a study
study = optuna.create_study(direction="maximize")

starttime = time.time()
study.optimize(objective, n_trials=200)
fittime = (time.time()-starttime)


print(f"Optimized R2: {study.best_value:.4f}")
print("Best params:")
for key, value in study.best_params.items():
    print(f"\t{key}: {value}")
```

```python
#retreiving best estimator


regressor = xgb.XGBRegressor(n_estimators=1000,**study.best_params)
regressor.fit(X_train,y_train) 

starttime = time.time()
y_predict = regressor.predict(X_test)
testtime = (time.time()-starttime) 

r2 = r2_score(y_test, y_predict)
rmse = np.sqrt(mean_squared_error(y_test, y_predict))
mae = median_absolute_error(y_test, y_predict)

new_row = pd.Series({'Model':"XGBoost", 'R2':r2, 'RMSE':rmse, 'MAE':mae,'Train_time': fittime, 'Test_time': testtime})
resultdf = pd.concat([resultdf, new_row.to_frame().T], ignore_index=True)
```

```python
# Get the feature importance scores
plt.figure(figsize=(10, 10))
sorted_idx = regressor.feature_importances_.argsort()[-25:]
height = regressor.feature_importances_[sorted_idx]
agg = np.flip(regressor.feature_importances_[sorted_idx]).cumsum()
bars = train.columns[sorted_idx]
y_pos = np.arange(len(bars))
 
plt.barh(y_pos, height)

plt.plot(np.flip(agg), bars,c="red",marker='o')

plt.yticks(y_pos, bars)
plt.grid()
 
plt.show()
```

```python
#plotting accuracy

plt.scatter(y_test, y_predict, c='r')
plt.xlabel('True values')
plt.ylabel('Predicted values')

xpoints = plt.xlim()
ypoints = plt.ylim()
plt.plot(xpoints, ypoints, 'k--', alpha=.8, scalex=False, scaley=False)

plt.grid()



plt.title('Model accuracy', fontsize=15)
plt.show()
```

As the best estimator we have is the XGBoost model we use this one to study the impact of the energy star score(we do it in the same fashion than before) :

```python
#splitting train and test values
etrain, etest = train_test_split(rawdf, test_size=0.2,random_state = splitrandomstate)

ey_train = etrain["TotalGHGEmissions"].values
ey_test = etest["TotalGHGEmissions"].values

etrain = etrain.drop(columns = ["SiteEnergyUseWN(kBtu)","TotalGHGEmissions"])
etest = etest.drop(columns = ["SiteEnergyUseWN(kBtu)","TotalGHGEmissions"])

eX_train = etrain.values
eX_test = etest.values
```

```python
edtrain = xgb.DMatrix(data = X_train, label = y_train)
edtest = xgb.DMatrix(data = X_test, label = y_test)

def eobjective(n_trials):
        
    params = {
        "learning_rate": n_trials.suggest_float("learning_rate", 0, 1),
        "min_split_loss": n_trials.suggest_float("min_split_loss", 0, 10),
        "reg_lambda": n_trials.suggest_float("reg_lambda", 0,1),
        "reg_alpha": n_trials.suggest_float("reg_alpha", 0,1),
        "max_depth": n_trials.suggest_int("max_depth", 2, 15),
        "colsample_bytree": n_trials.suggest_float("colsample_bytree", 0,1),
    }


    # Use cross-validation to calculate mean r2 score
    cv_results = xgb.cv(
        params, edtrain, num_boost_round=100, nfold=3, stratified=False, metrics="rmse",
        as_pandas=True, seed=123, early_stopping_rounds=50
    )
    
    r2 = -cv_results["test-rmse-mean"].iloc[-1]  # Negate rmse to maximize r2 (R^2 = 1 - (RMSE^2 / Var(y))where Var(y) is the variance of the true target values.)

    return r2
```

```python
# make a study
estudy = optuna.create_study(direction="maximize")

starttime = time.time()
estudy.optimize(eobjective, n_trials=200)
fittime = (time.time()-starttime)


print(f"Optimized R2: {estudy.best_value:.4f}")
print("Best params:")
for key, value in estudy.best_params.items():
    print(f"\t{key}: {value}")
```

```python
#retreiving best estimator


eregressor = xgb.XGBRegressor(n_estimators=1000,**estudy.best_params)
eregressor.fit(eX_train,ey_train) 

starttime = time.time()
ey_predict = eregressor.predict(eX_test)
testtime = (time.time()-starttime) 

r2 = r2_score(ey_test, ey_predict)
rmse = np.sqrt(mean_squared_error(ey_test, ey_predict))
mae = median_absolute_error(ey_test, ey_predict)

new_row = pd.Series({'Model':"eXGBoost", 'R2':r2, 'RMSE':rmse, 'MAE':mae,'Train_time': fittime, 'Test_time': testtime})
resultdf = pd.concat([resultdf, new_row.to_frame().T], ignore_index=True)
```

```python
# Get the feature importance scores
plt.figure(figsize=(10, 10))
sorted_idx = eregressor.feature_importances_.argsort()[-25:]
height = eregressor.feature_importances_[sorted_idx]
agg = np.flip(eregressor.feature_importances_[sorted_idx]).cumsum()
bars = etrain.columns[sorted_idx]
y_pos = np.arange(len(bars))
 
plt.barh(y_pos, height)

plt.plot(np.flip(agg), bars,c="red",marker='o')

plt.yticks(y_pos, bars)
plt.grid()
 
plt.show()
```
